```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/ultimate_student_productivity_dataset_5000.csv")

print("Dataset Shape:", df.shape)
df.head()
```

```python
# dataset info
df.info()

# statistical summary
df.describe()

# missing values
df.isnull().sum()
```

```python
plt.figure()
sns.histplot(df["study_hours"], bins=30)
plt.title("Distribution of Study Hours")
plt.xlabel("Study Hours")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
sns.countplot(x="gender", data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
df.groupby("academic_level")["exam_score"].mean().sort_values(ascending=False)
```

```python
df.sort_values("productivity_score", ascending=False).head(10)
```

```python
df.groupby("gender")["productivity_score"].mean()
```

```python
df.groupby("study_hours")["productivity_score"].mean().head()
```

```python
plt.figure(figsize=(10,8))

corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")

plt.show()
```

```python
plt.figure()

sns.scatterplot(
    x="study_hours",
    y="exam_score",
    data=df
)

plt.title("Study Hours vs Exam Score")
plt.show()
```

```python
plt.figure()

sns.scatterplot(
    x="sleep_hours",
    y="productivity_score",
    data=df
)

plt.title("Sleep Hours vs Productivity")
plt.show()
```

```python
plt.figure()

sns.boxplot(
    x="gender",
    y="productivity_score",
    data=df
)

plt.title("Productivity Score by Gender")
plt.show()
```

```python
plt.figure()

sns.boxplot(
    x="academic_level",
    y="productivity_score",
    data=df
)

plt.xticks(rotation=45)

plt.title("Productivity by Academic Level")
plt.show()
```

```python
df.groupby("gender")["exam_score"].mean()
df.groupby("gender")["exam_score"].mean().plot(kind="bar")
plt.title("Average Exam Score by Gender")
plt.show()
```

```python
top_students = df.sort_values(
    "exam_score",
    ascending=False
).head(10)

top_students
```
