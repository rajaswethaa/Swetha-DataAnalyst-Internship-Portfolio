```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("/content/drive/MyDrive/ultimate_student_productivity_dataset_5000.csv")
print("Dataset Shape:", df.shape)
df.head()
```

```python
avg_study = df["study_hours"].mean()
avg_exam = df["exam_score"].mean()
avg_productivity = df["productivity_score"].mean()
avg_sleep = df["sleep_hours"].mean()

print("Average Study Hours:", avg_study)
print("Average Exam Score:", avg_exam)
print("Average Productivity Score:", avg_productivity)
print("Average Sleep Hours:", avg_sleep)
```

```python
plt.figure()

sns.scatterplot(
    x="study_hours",
    y="exam_score",
    data=df
)

plt.title("Study Hours vs Exam Score")
plt.show()
```

```python
plt.figure()

sns.scatterplot(
    x="sleep_hours",
    y="productivity_score",
    data=df
)

plt.title("Sleep Hours vs Productivity")
plt.show()
```

```python
plt.figure()

sns.scatterplot(
    x="social_media_hours",
    y="productivity_score",
    data=df
)

plt.title("Social Media Usage vs Productivity")
plt.show()
```

```python
df.groupby("academic_level")["productivity_score"].mean()
```

```python
plt.figure()

sns.barplot(
    x="academic_level",
    y="productivity_score",
    data=df
)

plt.xticks(rotation=45)
plt.title("Productivity by Academic Level")

plt.show()
```

```python
plt.figure(figsize=(10,8))

corr = df.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True
)

plt.title("Correlation Matrix")
plt.show()
```

```python
from sklearn.cluster import KMeans

X = df[["study_hours","sleep_hours","social_media_hours"]]

kmeans = KMeans(n_clusters=3)

df["cluster"] = kmeans.fit_predict(X)

df.head()
```

```python
plt.figure()

sns.scatterplot(
    x="study_hours",
    y="exam_score",
    hue="cluster",
    data=df
)

plt.title("Student Segmentation")

plt.show()
```
